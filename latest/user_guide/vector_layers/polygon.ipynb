## Polygon

```python
m = folium.Map(location=[35.67, 139.78], zoom_start=13)

locations = [
    [35.6762, 139.7795],
    [35.6718, 139.7831],
    [35.6767, 139.7868],
    [35.6795, 139.7824],
    [35.6787, 139.7791],
]

folium.Polygon(
    locations=locations,
    color="blue",
    weight=6,
    fill_color="red",
    fill_opacity=0.5,
    fill=True,
    popup="Tokyo, Japan",
    tooltip="Click me!",
).add_to(m)

m
```
